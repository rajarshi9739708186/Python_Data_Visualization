```python
import matplotlib.pyplot as plt
```

# Plot a point

```python
x = 10  # X coordinate of the point
y = 20  # Y coordinate of the point
```

```python
plt.plot(x,y)  #Plotting X,Y. Not able to see Point. As we need to point it out with some charater
plt.xlabel("X-axis") #Name of X axis
plt.ylabel("Y-axis") #Name of Y axis
plt.title("Visualization of a point") # Title
plt.show() #Displaying the plot
```

```python
plt.plot(x,y, "o")  #Plotting X,Y. Pointed out with "o"
plt.xlabel("X-axis") #Name of X axis
plt.ylabel("Y-axis") #Name of Y axis
plt.title("Visualization of a point") # Title
plt.show() #Displaying the plot
```

```python
plt.plot(x,y, "x")  #Plotting X,Y. Pointed out with "x"
plt.xlabel("X-axis") #Name of X axis
plt.ylabel("Y-axis") #Name of Y axis
plt.title("Visualization of a point") # Title
plt.show() #Displaying the plot
```

# Set a figure size

#### function call of plt.figure(figsize=(x,y)) should be before plt.plot(). Otherwise plotting will be done of default size and a new seperate figure will be displayed

```python
plt.figure(figsize=(8,8)) # Setting Figure size
plt.plot(x,y, "+")  #Plotting X,Y. Pointed out with "+"
plt.xlabel("X-axis") #Name of X axis
plt.ylabel("Y-axis") #Name of Y axis
plt.title("Visualization of a point") # Title
plt.show() #Displaying the plot
```

```python
plt.plot(x,y, "+")  #Plotting X,Y. Pointed out with "+"
plt.figure(figsize=(8,8)) # Setting Figure size after plotting. It will create seperate figure
plt.xlabel("X-axis") #Name of X axis
plt.ylabel("Y-axis") #Name of Y axis
plt.title("Visualization of a point") # Title
plt.show() #Displaying the plot
```

# Multiple Plots

```python
plt.plot(2,2, "o")  #Plotting X,Y. Pointed out with "o"
plt.plot(6,7, "o")  #Plotting X,Y. Pointed out with "o"
plt.plot(10,11, "o")  #Plotting X,Y. Pointed out with "o"
plt.xlabel("X-axis") #Name of X axis
plt.ylabel("Y-axis") #Name of Y axis
plt.title("Visualization of a point") # Title
plt.show() #Displaying the plot
```

```python
for value in range(1,5):
    plt.plot(value, value, "o")
plt.xlabel("X-axis") #Name of X axis
plt.ylabel("Y-axis") #Name of Y axis
plt.title("Visualization of a point") # Title
plt.show() #Displaying the plot
```
